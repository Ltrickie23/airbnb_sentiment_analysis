# LLM - DistilBERT

DistilBERT is a lighter version of BERT that retains 97% of its performance at 40%. It was pre-trained on large text and fine-tuned specifically for sentiment analysis, making it ready to use without any additional training.

```python
import pandas as pd
import numpy as np
from transformers import pipeline
from sklearn.metrics import classification_report

df = pd.read_csv(r"C:\Users\kumri\Downloads\Data Science Notebooks\AirBnB sentiment analysis\sentiment_tags.csv")
print(df.shape)
```

```python
classifier = pipeline('text-classification',
                      model='distilbert-base-uncased-finetuned-sst-2-english')
#examples before running all the data.
print(classifier("Amazing place, I loved every minute of it!"))
print(classifier("Terrible place, would not recommend at all."))
print(classifier("The place was okay, nothing special."))
```

## Binary classification only

DistilBERT was fine-tuned on SST-2, which only has 2 classes: Positive and Negative. This makes it unsuitable for our three-class use case.

```python
print(classifier.model.config.id2label)
```

**Conclusion:**
Linear SVC was selected as the final model. It achieves the highest overall accuracy (91%) and leads or ties on F1-score across all three sentiment classes — particularly on negative review detection (94%), which is critical for identifying properties with poor guest experiences.
